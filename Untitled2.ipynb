```python
import pandas as pd
```

```python
import glob
```

```python
import bin
import bin.utils

aime24df = bin.utils.load_dataset('aime24')
```

```python
df = pd.concat([pd.read_csv(filepath, index_col=0).assign(id=int(filepath.split("/")[-1].split('.')[0])) for filepath in sorted(glob.glob("resources/gpt4omini-aime24-ac/*"))]).reset_index(drop=True)
df2 = aime24df.target.to_frame()
df2['answer'] = df.groupby('id').answer.agg(lambda x: x.value_counts().index[0])
df2['trials'] = df.groupby('id').token_usage.count()
df2['token_usage'] = df.groupby('id').token_usage.sum()
df2
```
