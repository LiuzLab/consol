```python
def self_consistency(prompt, OutputFormat, trials=40):
    _ = llm.with_structured_output(OutputFormat, include_raw=True).batch([[
        langchain_core.messages.HumanMessage(prompt),
    ]] * trials)
    raw_outputs = []
    while len(raw_outputs) < trials:
        try:
            k = trials - len(raw_outputs)
            _ = llm.with_structured_output(OutputFormat, include_raw=True).batch([[
                langchain_core.messages.HumanMessage(prompt),
            ]] * k)
            raw_outputs += [x for x in _ if x['parsed']]
        except:
            continue
    ss, token_usage = pd.Series([x['parsed'].answer for x in raw_outputs]).value_counts(), pd.Series([x['raw'].response_metadata['token_usage']['completion_tokens'] for x in raw_outputs]).sum()

    return ss.index[ss.argmax()], 40, token_usage
self_consistency(df.input.iloc[2], ReasonedOutput)
```

```python
def get_opportunistic_trials(first, second):
    assert first >= second
    for k in range(0, 40):
        if first+k == 0:
            continue
        if scipy.stats.binomtest(first+k, first+second+k, p=0.5, alternative='greater').pvalue <= 0.05:
            return k
    raise "Opportunistic Trials Not Found"

def adaptive_consistency(prompt, OutputFormat, max_trials=40):
    total_ss = pd.Series()
    total_trials = 0
    total_token_usage = 0
    with tqdm.notebook.tqdm(leave=False, total=max_trials) as pbar:
        while True:
            two = total_ss.sort_values(ascending=False).head(2).to_list()
            while len(two) < 2:
                two += [0]
            trials = get_opportunistic_trials(two[0], two[1])
            if trials >= max_trials - total_trials:
                trials = max_trials - total_trials

            if trials == 0:
                break

            raw_outputs = []
            while len(raw_outputs) < trials:
                try:
                    k = trials - len(raw_outputs)
                    _ = llm.with_structured_output(OutputFormat, include_raw=True).batch([[
                        langchain_core.messages.HumanMessage(prompt),
                    ]] * k)
                    raw_outputs += [x for x in _ if x['parsed']]
                except:
                    print("unknown error")
                    continue
            ss, token_usage = pd.Series([x['parsed'].answer for x in raw_outputs]).value_counts(), pd.Series([x['raw'].response_metadata['token_usage']['completion_tokens'] for x in raw_outputs]).sum()

            total_ss = total_ss.add(ss, fill_value=0).astype('int')
            total_trials += trials
            total_token_usage += token_usage
            pbar.update(trials)

    return total_ss.index[total_ss.argmax()], total_trials, total_token_usage

_ = adaptive_consistency(df.input.iloc[2], ReasonedOutput)
_
```
